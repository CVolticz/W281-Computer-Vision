```python
def main(path_in, path_out):
  
    if not os.path.exists(path_out):
        os.makedirs(path_out)

    print('Start augmenting images')

    datagen = ImageDataGenerator(
        rotation_range = 20,
        shear_range = 0.05,
        zoom_range = 0.05,
        horizontal_flip = False,
        brightness_range = (0.75, 1.25)
      )

    samples_per_img = 5
    
    image_files = os.listdir(path_in)
    for i, image_file in tqdm(enumerate(image_files)):
      img = load_img(os.path.join(path_in, image_file))
      x = img_to_array(img)
      x = x.reshape((1, ) + x.shape)
      j = 0
      image_file_no_type = image_file.split(".")[0]
      for batch in datagen.flow(x, batch_size = 1, save_to_dir = path_out, save_prefix = f'augmented_{image_file_no_type}', save_format ='jpeg'):
        j += 1
        if j >= samples_per_img:
            break

path_in = f'{DATA_DIR}/raw/Coronahack-Chest-XRay-Dataset/Coronahack-Chest-XRay-Dataset/train_resized'
path_out = f'{DATA_DIR}/raw/Coronahack-Chest-XRay-Dataset/Coronahack-Chest-XRay-Dataset/train_resized_augmented'
main(path_in, path_out)
```
