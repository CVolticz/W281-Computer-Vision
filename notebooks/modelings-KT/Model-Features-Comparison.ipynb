```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# LBP Feature Measurements

```python
data = {
    "Logistic Regression": 0.39,
    "SVC": 0.39,
    "GBC": 0.63,
    "autoML": 0.65
}

accuracy_lbp = pd.DataFrame().from_dict(data, orient='index')
accuracy_lbp
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,6))
accuracy_lbp[0].plot(kind='bar')
# set figure title

fig.suptitle('LBP Feature Models (PCA) Test Accuracy', fontsize=20)
# set x and y labels
ax.set_xlabel('Model', fontsize=14)
ax.set_ylabel('Accuracy', fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(rotation = 0, fontsize=14)
```
